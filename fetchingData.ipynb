```python
# until now we only fetched the data form the 0th number of the json dictionary 
# to fetch overall data we need to run a for loop
```

```python
df = pd.DataFrame()
```

```python
df # this is an empty dataFrame
```

```python
for i in range (1, 500):
    response = rs.get('https://api.themoviedb.org/3/tv/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','name','first_air_date','origin_country','original_language','overview','popularity','vote_average','vote_count']]
    df = df.append(temp_df,ignore_index=True)
```
